```python
nuc2vec={}
for i1,N1 in enumerate(["A","C","G","T"]):
        nuc2vec[N1]=i1

def seq2vec(seq):
    L=len(seq)
    vec=np.zeros((L,4))
    for i in range(L):
        j=nuc2vec[seq[i]]
        vec[i,j]=1
    return vec

def seq2vec2(seq):
    L=len(seq)
    vec=np.zeros((L,2))
    for i in range(L):
        j=nuc2vec[seq[i]]
        if j==0:   vec[i,0]=0; vec[i,1]=1
        elif j==1: vec[i,0]=1; vec[i,1]=0
        elif j==2: vec[i,0]=1; vec[i,1]=1
        elif j==3: vec[i,0]=0; vec[i,1]=0
    return vec

def vec2seq(vec):
    seq_argmax = np.argmax(vec[:,:4],axis=1)
    seq = ''
    for j in range(47):
        if seq_argmax[j] == 0: seq += 'A' 
        elif seq_argmax[j] == 1: seq += 'C' 
        elif seq_argmax[j] == 2: seq += 'G' 
        elif seq_argmax[j] == 3: seq += 'T' 
        else: print('error in result2seq')
    return seq

def log_mse_loss(y_true,y_pred):
    log_mse = K.mean(K.square(K.log(y_pred+1)-K.log(y_true+1)),axis=-1)
    return log_mse

def mse_loss(y_true,y_pred):       
    mse = K.mean(K.square((y_pred+1)-(y_true+1)),axis=-1)
    return mse
```

```python
###Construct a CNN model - divide into small regions+convolutional filter - without Reshape

filterN = 10
filterL = 8

coef_array_ave = np.zeros((filterN,1,47,5))

maxlen=47
input_shape = (1,maxlen,5)
inp = Input(shape=input_shape)
drop1 = Dropout(0.2)(inp)

region = {}
conv2 = {}
for k in range(47-filterL+1):
    region[k] = keras.layers.Cropping2D(cropping=((0,0),(k,47-filterL-k)))(drop1)
    conv2[k] = Conv2D(filterN,kernel_size=(filterL,5),strides=(1,5), \
                      activation='relu',padding='valid', data_format="channels_first")(region[k])

merged = Concatenate(axis=-2)([conv2[k] for k in range(47-filterL+1)])
drop2 = Dropout(0.2)(merged)
flt2 = Flatten()(drop2)
dns = Dense(10, activation='relu')(flt2)
outp = Dense(1, activation='relu')(dns)
model = Model(inputs=inp,outputs=outp)

model.summary()
```

```python
tmpDF = wholeDF
seqs = list(set(tmpDF.iloc[:,2]))
[train_seqs, vald_seqs, test_seqs] = pickle.load(open(filepath+"index.pkl","rb"))

ERdf = {}
for seq in seqs:
    ERdf[seq] = tmpDF[tmpDF.iloc[:,2]==seq][["PBS length", "RT length", "Measured PE efficiency"]]
seq_arrays = {}
ERs = {}
data_lst = {}
for S, seqs in zip(["train","test","vald"],[train_seqs, test_seqs, vald_seqs]):
    seq_arrays[S] = []
    ERs[S] = []
    for seq in seqs:
        seq_array = seq2vec(seq).reshape((1,47,4))
        for PBS, RT, ER in zip(ERdf[seq]["PBS length"], ERdf[seq]["RT length"], ERdf[seq]["Measured PE efficiency"]):
            indicator = np.zeros((1,47,1))
            indicator[:,21-PBS:21+RT,:]=1
            seq_arrays[S].append(np.concatenate([seq_array,indicator], axis=2))
            ERs[S].append(ER)
    seq_arrays[S] = np.concatenate(seq_arrays[S])
    seq_arrays[S] = seq_arrays[S].reshape((-1,1,47,5))
    ERs[S] = np.array(ERs[S]).reshape((-1,1))
    ERs[S] = np.where(ERs[S]>0, ERs[S], 0)
    data_lst[S] = [seq_arrays[S],ERs[S]]
```

```python
wholeDF = pd.read_csv("./NatureBiotech/ER_PBS_RT.csv",skiprows=0, header=1)

ERDF = {}
for PBSL in list(set(wholeDF["PBS length"])):
    ERDF[PBSL] = wholeDF[wholeDF["PBS length"]==PBSL].copy()

guideSeq = {}
for PBSL in list(set(wholeDF["PBS length"])):
    for RTTL in [10,12,15,20]:
        guideSeq[PBSL, RTTL] = set(ERDF[PBSL][(ERDF[PBSL]["RT length"]==RTTL)]["Guide sequence ((G/g)N19)"])

seq2stay = {}
for PBSL in list(set(wholeDF["PBS length"])):
    seq2stay[PBSL] = guideSeq[PBSL,10]
    for RTTL in [12,15,20]:
        seq2stay[PBSL] = seq2stay[PBSL].intersection(guideSeq[PBSL,RTTL])
        
for PBSL in list(set(wholeDF["PBS length"])):
    bool_col = []
    for seq in ERDF[PBSL]["Guide sequence ((G/g)N19)"]:
        if seq in seq2stay[PBSL]:
            bool_col.append(1)
        else:
            bool_col.append(0)
    ERDF[PBSL]["bool"] = bool_col
    ERDF[PBSL] = ERDF[PBSL][ERDF[PBSL]["bool"]==1]
    ERDF[PBSL]=ERDF[PBSL].sort_values(by=["Guide sequence ((G/g)N19)","RT length"])
    
simpDF = {}
for PBSL in list(set(wholeDF["PBS length"])):
    lst = []
    for row in ERDF[PBSL].itertuples():
        if row[7]==10:
            tmp = [row[-2]]
        elif row[7]==12:
            tmp.append(row[-2])
        elif row[7]==15:
            tmp.append(row[-2])
        elif row[7]==20:
            tmp.append(row[-2])
            tmp.append(row[3])
            lst.append(tmp)
    simpDF[PBSL] = pd.DataFrame(lst)
    simpDF[PBSL].columns=["ER10", "ER12", "ER15", "ER20", "seq"]
    
for PBSL in list(set(wholeDF["PBS length"])):
    simpDF[PBSL]["delER12_10"]=simpDF[PBSL]["ER12"]-simpDF[PBSL]["ER10"]
    simpDF[PBSL]["delER15_10"]=simpDF[PBSL]["ER15"]-simpDF[PBSL]["ER10"]
    simpDF[PBSL]["delER20_10"]=simpDF[PBSL]["ER20"]-simpDF[PBSL]["ER10"]
    simpDF[PBSL]["delER15_12"]=simpDF[PBSL]["ER15"]-simpDF[PBSL]["ER12"]
    simpDF[PBSL]["delER20_15"]=simpDF[PBSL]["ER20"]-simpDF[PBSL]["ER15"]
    simpDF[PBSL]["delER20_12"]=simpDF[PBSL]["ER20"]-simpDF[PBSL]["ER12"]
```

```python
### max sequence
pathname = "/home/groups/song/songlab2/dnewton2/FinalSim/cutoff_25/PBSL13_RTTL15/"
chains = [f for f in listdir(pathname) if isfile(join(pathname, f))]
chains = sorted([f for f in chains if (".csv" in f)])
last_seqs = []
for chain in chains:
    with open(pathname+chain) as file:
        maxER = 0
        maxseq = ""
        for i,line in enumerate(file):
            currentseq = line.split(",")[-2]
            currentER = line.split(",")[-1].strip("\n")
            if currentseq == "Sequence": continue
            if float(currentER) > float(maxER): 
                maxi = i
                maxER = currentER
                maxseq = currentseq
    print("%6d" % maxi, "%.3f" % float(maxER), maxseq)
    seq_array = seq2vec(maxseq).reshape((1,47,4))
    PBS = 13
    RT = 15
    indicator = np.zeros((1,47,1))
    indicator[:,21-PBS:21+RT,:]=1
    last_seqs.append(np.concatenate([seq_array,indicator], axis=2))
last_seqs = np.array(last_seqs)
```
